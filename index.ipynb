### Build normalized histograms with kde plots to explore the distributions further. 
### Use only the continuous variables in the dataset to plot these visualizations. 

```python
#Your code here 
for col in ['CPI','Dept', 'Fuel_Price',  'Size', 'Store', 'Temperature','Unemployment', 'Weekly_Sales']:
    df[col].hist(density=True, label=col)
    df[col].plot.kde(label=col)
    plt.legend()
    plt.show()
```

```python
# State your observations here 
# Most of the graphs still show little sign of normaility.  The store, fuel_price, and density graphs are multimodal.
```

### Build joint plots to check for the linearity assumption between predictors and target variable

Let's use a slightly more advanced plotting technique in seaborn that uses scatter plots, distributions, kde and simple regression line - all in a single go. Its called a `jointplot`. [Here is the official doc. for this method](https://seaborn.pydata.org/generated/seaborn.jointplot.html). 

Here is how you would use it:

> **`sns.jointplot(x= <column>, y= <column>, data=<dataset>, kind='reg')`**

A joint plot will allow us to visually inspect linearity as well as normality assumptions as a single step. 

```python
#Your code here 
plt.style.use('seaborn')

for col in ['CPI','Dept', 'Fuel_Price',  'Size', 'Store', 'Temperature','Unemployment']:
    sns.jointplot(x= col, y= 'Weekly_Sales', data=df, kind='reg', label=col)
    plt.legend()
    plt.show()
```

```python
#Provide your observations here 
# The regression lines show a few weak relationships. Size has a positive slope, CPI has a negative slope.
# The other plots show no visible relationship with an almost horizontal regression line across the y-intercept.
# Plots have sales values evenly uniformly distributed down the x-axis showing no clear relationship for most predictors.
# KDEs still show little normality.  
# Ultimately not an overall impressive set of predictors, may need more processing.
```
